# Harish Practise Project: Rock vs Mine Prediction

This ML project uses sonar signal data to determine if the detected object is a **rock** or a **mine**. Each object has 60 sonar signal readings (numerical features), and the target label is either `'R'` (Rock) or `'M'` (Mine).


# Importing Required Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

# Loading the Dataset

```python
df = pd.read_csv('sonar data.csv',header = None)
print("Dataset Shape:", df.shape)
df.head()
```

# Another way : Number of rows and column

```python
df.shape
```

# Describe --> statistical measures of the data

```python
df.describe()
```

```python
print("Label distribution:\n", df[60].value_counts())
```

# Mean of each feature by class

```python
df.groupby(60).mean()
```

# Separating data and Labels

```python
X = df.drop(columns=60, axis=1)
Y = df[60]
print(X)
print(Y)
```

# Data Preprocessing

```python
X = df.drop(columns = 60)
Y = df[60]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify = Y, random_state = 1)
print("Train shape :", X_train.shape)
print("Test shape :", X_test.shape)
```

# Model Training --> Logistic Regression

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

# Accuracy Evaluation

```python
train_pred = model.predict(X_train) #pred = prediction
train_acc = accuracy_score(Y_train, train_pred) #acc = accuracy
print("Training Accuracy :", train_acc)
test_pred = model.predict(X_test)
test_acc = accuracy_score(Y_test, test_pred)
print("Testing Accuracy :", test_acc)
```

# Confusion Matrix & Report

```python
cm = confusion_matrix(Y_test, test_pred)
sns.heatmap(cm, annot=True, fmt = 'd', cmap = "Blues")
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
print("Classification Report :\n", classification_report(Y_test, test_pred))
```

# Prediction System with Interactive Input

```python
# Interactive Input: User enters 60 comma-separated values
input_str = input("Enter 60 sonar readings separated by commas:\n")
input_data = tuple(map(float, input_str.strip().split(',')))

# Convert and reshape
input_array = np.asarray(input_data).reshape(1, -1)

# Predict
result = model.predict(input_array)[0]

# Output
if result == 'R':
    print("ðŸ”µ The object is a Rock.")
else:
    print("ðŸ”´ The object is a Mine.")
```
